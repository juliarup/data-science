### Usuwanie białych znaków
Nazwy atrybutów nie są znormalizowane, posiadają przerwy na początku i na końcu

```python
# Wyświetl oryginalne nazwy kolumn
print("Oryginalne nazwy kolumn:")
print(df.columns)

# Zmiana nazw kolumn poprzez usuwanie białych znaków
df.rename(columns=lambda x: x.strip(), inplace=True)

df = df.rename(columns={'x dimension': 'x_dimension',
                        'y dimension': 'y_dimension',
                        'z dimension': 'z_dimension'
                        })

# Wyświetl zaktualizowane nazwy kolumn
print("\nZaktualizowane nazwy kolumn:")
print(df.columns)
```

### Sprawdzenie duplikatów różnymi sposobami, np. po wybranych atrybutach

`.duplicated()` skorzystać z dwóch argumentów: `subset` i `keep`. Dla argumentu subset podaj listę nazw kolumn, w których chcemy sprawdzić duplikaty, a w argumencie keep podaj  1 z 3 wartości: „pierwsza”, „ostatnia” lub „Fałsz”. Z dokumentacji wiemy, że:
- `first` : Oznacz duplikaty jako True z wyjątkiem pierwszego wystąpienia.
- `last` : Zaznacz duplikaty jako True z wyjątkiem ostatniego wystąpienia.
- `False` : Oznacz wszystkie duplikaty jako True.

```python
duplicates = df.duplicated (subset=["depth", "table", "price"], keep=False)

df[duplicates]
```

```python
## Posortuj zduplikowane wartości według odpowiedniego indeksu
df[duplicates].sort_values("price")
```

```python
duplicates2 = df.duplicated (subset=["carat", "color", "price"], keep=False)

df[duplicates2].sort_values("carat")
```

## Brakujące wartości
Brakujące wartości typu float zostaną zamienione na średnią w danej kolumnie, pozostałe wartości zostaną uzupełnione modą

```python
columns = df.columns.to_list()

for column in columns:
    if  isinstance(df[column], float):
        df[column].fillna(df[column].mean(), inplace = True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)

"""df['carat'].fillna(df['carat'].mean(), inplace=True) # średnia
df['x dimension'].fillna(df['x dimension'].mean(), inplace=True)
df['y dimension'].fillna(df['y dimension'].mean(), inplace=True)
df['z dimension'].fillna(df['z dimension'].mean(), inplace=True)
df['depth'].fillna(df['depth'].mean(), inplace=True)
df['table'].fillna(df['table'].mean(), inplace=True)
df['price'].fillna(df['price'].mean(), inplace=True)"""
```
